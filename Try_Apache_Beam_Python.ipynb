SETUP

```python
# ===========================
# SETUP CELL
# ===========================

import apache_beam as beam
import re

INPUT_FILE = 'world_events_dataset.txt'
OUTPUT_PREFIX = 'outputs'   
```

Word Count With Comments

```python
# ===========================
# WORD COUNT CELL
# ===========================

with beam.Pipeline() as p:

    lines = (
        p
        | "ReadInputFile" >> beam.io.ReadFromText(INPUT_FILE)
    )

    words = (
        lines
        | "ExtractWords" >> beam.FlatMap(
            lambda line: re.findall(r"[A-Za-z']+", line)
        )
    )

    word_counts = (
        words
        | "CountWordsWithPerElement" >> beam.combiners.Count.PerElement()
    )

    sorted_word_counts = (
        word_counts
        | "ToList" >> beam.combiners.ToList()
        | "SortByCountDesc" >> beam.FlatMap(
            lambda items: sorted(items, key=lambda kv: kv[1], reverse=True)
        )
    )

  
    formatted = (
        sorted_word_counts
        | "FormatAsTuple" >> beam.Map(lambda kv: (kv[0], kv[1]))
    )

    formatted | "PrintResults" >> beam.Map(print)

    formatted_strings = (
        sorted_word_counts
        | "FormatAsString" >> beam.Map(lambda kv: f"('{kv[0]}', {kv[1]})")
    )

    _ = (
        formatted_strings
        | "WriteToText" >> beam.io.WriteToText(
            OUTPUT_PREFIX,
            file_name_suffix='.txt'
        )
    )
```
